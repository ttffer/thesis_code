```python
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
#stuck at 1
figure(figsize=(8, 6), dpi=100)
uncode_infected = np.array([90.9,90.4,89.5,87.2,87,87,87,82.9])
encoded_infected = np.array([85.5,85.4,85.3,84.5,84,84,84,77.6])
encoded_infected_with_decoder = np.array([85.5,85.5,85.5,85.5,85.5,85.5,85.5,85.5,])
xpoints = np.arange(21,29)


plt.plot(xpoints, uncode_infected,'-o',label='uncode_infected')
plt.plot(xpoints, encoded_infected,'-o',label='encoded_infected')
plt.plot(xpoints, encoded_infected_with_decoder,'-o',\
         label='encoded_infected_with_decoder')
plt.xlabel('Bit position')
plt.ylabel('Accuracy (%)')
plt.grid()
plt.title("Accuracy Result When Any Error Index Stuck at 1")
plt.legend()
```
